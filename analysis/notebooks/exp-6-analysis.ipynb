```python
import pandas as pd

def clean_data(df):
    # Drop columns: 'data', 'epochs' and 12 other columns
    df = df.drop(columns=['experiment_name', 'add_start_end_tokens', 'add_row_id', 'add_label_tokens', 'train_data_percent_used', 'tokenization', 'epochs', 'batch_size', 'learning_rate', 'seq_length', 'test_split', 'val_split', 'vocab_size', 'patience', 'results_dir', 'shuffle', 'disable_train_shuffle', 'early_stopping', 'remove_system_tables'])
    # Sort by column: 'actual_test_accuracy_mean' (descending)
    df = df.sort_values(['horizon'], ascending=[True])
    return df

df = pd.read_csv('../results/exp-6/results_clean.csv')
df_clean = clean_data(df.copy())

df_longer = pd.read_csv('../results/exp-6-longer/results_clean.csv')
df_clean_longer = clean_data(df_longer.copy())
```

```python
on=['model', 'data', 'horizon', 'token_length_seq']
metrics = ['actual_test_accuracy_mean', 'actual_test_accuracy_std']
drop = [x for x in df_clean.columns if x not in on + metrics]
rename = lambda y: {x: x + f'_{y}' for x in metrics}
df_clean = df_clean.drop(columns=drop).rename(columns=rename(50))
df_clean_longer = df_clean_longer.drop(columns=drop).rename(columns=rename(100))
df_joined = df_clean.merge(df_clean_longer, on=on, how='inner')
df_joined

```
